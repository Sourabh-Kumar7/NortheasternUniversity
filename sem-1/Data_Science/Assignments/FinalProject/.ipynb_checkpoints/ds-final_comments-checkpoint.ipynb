```python
from scipy.stats import pointbiserialr

correlation_results = {}

for column in clear_df.columns:

    if column != 'Mortality':
        correlation, p_value = pointbiserialr(clear_df[column], clear_df['Mortality'])
        
        correlation_results[column] = {'correlation': correlation, 'p_value': p_value}

correlation_df = pd.DataFrame.from_dict(correlation_results, orient='index')

print(correlation_df.sort_values(by='correlation'))
```

```python
from scipy.stats import chi2_contingency

correlation_results = {}

for column in clear_df.columns:

    if column != 'Mortality':
        contingency_table = pd.crosstab(clear_df[column], clear_df['Mortality'])
        chi2, p_value, dof, expected = chi2_contingency(contingency_table)
        
        correlation_results[column] = {'chi2': chi2, 'p_value': p_value, 'dof': dof, 'expected': expected}

correlation_df = pd.DataFrame.from_dict(correlation_results, orient='index')

plt.figure(figsize=(10, 6))
sns.barplot(x=correlation_df['chi2'], y=correlation_df.index)
plt.title('Chi-squared Values of Features')
plt.xlabel('Chi-squared Value')
plt.ylabel('Features')
plt.show()

correlation_df['neg_log_pval'] = -np.log10(correlation_df['p_value'])

plt.figure(figsize=(10, 6)) 
sns.barplot(x=correlation_df['neg_log_pval'], y=correlation_df.index)
plt.title('Negative Log P-values of Features')
plt.xlabel('-Log10(P-value)')
plt.ylabel('Features')
plt.show()
```

```python
important_features = ['Age', 'chol', 'ca', 'thalach', 'trestbps', 'Follow.Up', 'oldpeak', 'Diabetes', 'Hemoglobin', 'Reaction', 'Mortality']
columns_to_drop = [col for col in clear_df.columns if col not in important_features]

model_df_standard = clear_df.drop(columns=columns_to_drop)
```

```python
binary_columns = ['Diabetes', 'Reaction', 'Mortality']

binary_features = model_df_standard[binary_columns]
continuous_features = model_df_standard.drop(columns=binary_columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_continuous_features = scaler.fit_transform(continuous_features)

scaled_continuous_features_df = pd.DataFrame(scaled_continuous_features, columns=continuous_features.columns, index=continuous_features.index)
```

```python
scaled_df = pd.merge(scaled_continuous_features_df, binary_features, left_index=True, right_index=True)
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split

test_size = 0.25
train_df, test_df = train_test_split(scaled_df, test_size=test_size, random_state=RANDOM_SEED)
```

```python
from imblearn.over_sampling import SMOTE

X = train_df.drop('Mortality', axis=1)
y = train_df['Mortality']

smote = SMOTE(random_state=RANDOM_SEED)

X, y = smote.fit_resample(X, y)
```

Bayes theorem says

$P(model/data) = (P(data/model) * P(model))  / P(data)$

which is also 

$P(model/data) \propto P(data/model) * P(model)$

P(model) is the prior

P(data/model) is the likelihood

P(model/data) is the posterior

##### Model equation for multiple logistic regression #
$mu = alpha + beta_1 * data['Feature1'] + beta_2 * data['Feature2'] + ... + beta_n * data['FeatureN']$
<br/>
$μ=α+β1⋅x1+β2⋅x2+…+βn⋅xn$

The training data is inserted into the pymc model as mutable to change it later to testing data.
 
The priors were chosen to be drawn from a Normal distribution with a mean of 0 and a variance of 10. This was due to an initial assumption about most of the features, in particular, about them being normally distributed. This was later confirmed by the posterior alpha and betas, since they were all observed to be around the variance of the prior.

The model uses logistic regression sigmoid function, which combines one alpha intercept and betas for each of the features to create a function which deviates from 0 to 1. Due to the binary nature of the target outcome, the Bernoulli is used as the likelihood function which accepts the logistic regression function as its mu.

The trace results from MCMC sampling using NUTS after 20,000 samples over 4 chains.

```python
X_test = test_df.drop('Mortality', axis=1)
y_test = test_df['Mortality']

with pm.Model() as logistic_model:
    X_shared = pm.MutableData('X_shared', X)
    y_shared = pm.MutableData('y_shared', y)
    
    alpha = pm.Normal('alpha', mu=0, sigma=10)
    betas = pm.Normal('betas', mu=0, sigma=10, shape=X_shared.shape[1])

    logits = alpha + pm.math.dot(X_shared, betas)

    likelihood = pm.Bernoulli('y', pm.math.sigmoid(logits), observed=y_shared)

    trace = pm.sample(20000, tune=1000, target_accept=0.80, return_inferencedata=True, chains=4)
```

```python
gv = pm.model_graph.model_to_graphviz(logistic_model)
display_png(gv)
```

```python
print(f"Number of features in Test Data: {X_test.shape[1]}")
print(f"Number of features in Train Data: {X.shape[1]}")
print(f"Chains in trace: {len(trace)}")
```

```python
import arviz as az

summary = az.summary(trace)
print(summary)

az.plot_trace(trace)
plt.show()

az.plot_posterior(trace)
plt.show()
```

We can observe the trace plots for each of the 10 chosen features as well as the burn-in period which is pretty much non-existent due to NUTS nature.

A number of posterior betas are close to each other, all the features are also normally distributed. Some of the resulting posterior betas are negative.

### Prediction #
The following prediction uses a mathematical approach, as well as generalizes the model's posterior results in order to make predictions based on unseen, test data.

One of the caveats of the test data is that the two classes (mortal cases, and not) are imbalanced. Using over-sampling techniques, such as SMOTE which was used on training data, on the test data, would result in data leakage. 

This approach to predictions is the most appropriate one given the nature of the dataset. The goal is to predict the mortality of an individual by considering an average scenario. This generalized prediction does not capture all the uncertainties of the underlying model. 

```python
alpha_samples = trace.posterior['alpha'].values
betas_samples = trace.posterior['betas'].values

print("Alpha samples shape:", alpha_samples.shape)
print("Betas samples shape:", betas_samples.shape)
```

In the next cell, the mean values for each of the feature betas are considered as well as a mean alpha.

The test data is then entered individually into the logistic regression function and an arbitrary cutoff of 0.5 is used to predict the results given feature data.

```python
alpha_mean = np.mean(alpha_samples, axis=(0, 1)) 
betas_mean = np.mean(betas_samples, axis=(0, 1)) 

print("X_test shape:", X_test.shape)
print("Any NaNs in X_test:", np.any(np.isnan(X_test)))
print("Any infinities in X_test:", np.any(np.isinf(X_test)))

logits = alpha_mean + np.dot(X_test, betas_mean.T)
prob_predictions = 1 / (1 + np.exp(-logits))

binary_predictions = (prob_predictions > 0.5).astype(int)
```

```python
binary_predictions
```

### Model Accuracy #

The accuracy of this approach is quite good, given the constraints of the dataset as well as those of Bayesian Estimation.

The model resulted in the accuracy score of 76%. A more important score, ROC-AUC, which accounts for the discrepancy in class sizes in the test data was also determined to be 77%.
The false positive precision score is quite low, at 55%. However, the recall score which is concerned with the proportion of actual positives that were correctly identified, also ended up being quite high at 77%. Recall score is particularly useful since the cost of false negatives is high given this dataset.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score

accuracy = accuracy_score(y_test, binary_predictions)
precision = precision_score(y_test, binary_predictions)
recall = recall_score(y_test, binary_predictions)
roc_auc = roc_auc_score(y_test, prob_predictions)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"ROC-AUC: {roc_auc}")
```

### Sample posterior predictive #

This prediction method uses pymc sample_posterior_predictive function which draws samples from the posterior in order to make predictions about the given data. This method captures all the uncertainties of the model, as well as follows the stochastic Bayesian estimation approach. However, the resulting predictions were not anywhere near appropriate.

This is due to the function drawing from the entire sample space of the posterior, which means that certain beta and alpha values are far from average. Therefore, the resulting predictions are all around 0.5. There are only two potential outcomes: 0 and 1. Out of 20,000 samples per chain (4 chains in total), there are many predictions of 1 and many predictions of 0. This results in a mean prediction of approximately 0.5 for most of the test data.

Considering the code, first the data is switched inside the model from training to testing, unseen data. Then the sample_posterior_predictive function is called. The trace is a combination of 20,000 predictions for each test data. The mean of them along the two axis results in the average prediction value.

```python
with logistic_model:
    pm.set_data({"X_shared": X_test})
    pm.set_data({"y_shared": y_test})
    
    posterior_pred_test = pm.sample_posterior_predictive(trace, random_seed=RANDOM_SEED)
    
prob_predictions = np.mean(posterior_pred_test.posterior_predictive['y'].values, axis=(0, 1))
```

```python
plt.hist(prob_predictions)
```

Choosing an arbitrary threshold value by looking at the distribution of the data results in data leakage since we assume the distribution of the test data mortality is the same with the entire dataset which might not be the case since the data is split at random.

```python
threshold = 0.51
binary_predictions = (prob_predictions > threshold).astype(int)

if len(binary_predictions) != len(y_test):
    raise ValueError("Mismatch in the length of predictions and actual values.")
```

```python
binary_predictions
```

In this case, the results are not exemplary. The accuracy score is 43%, precision at about 22%, recall at 38%, and ROC-AUC at 43%.

```python
print(len(y_test), len(X_test), len(binary_predictions))

accuracy = accuracy_score(y_test, binary_predictions)
precision = precision_score(y_test, binary_predictions)
recall = recall_score(y_test, binary_predictions)
roc_auc = roc_auc_score(y_test, prob_predictions)  # Use probabilities for ROC-AUC

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"ROC-AUC: {roc_auc}")
```

The dataset was modeled using Bayesian estimation with the logistic regression model used to combine the features and Bernoulli as the likelihood function. The model was trained on a set of features selected based on their linear and non-linear correlation with the target variable 'Mortality'. The model's performance was evaluated using accuracy, precision, recall, and ROC-AUC scores.

The model achieved an accuracy of 76%, indicating that it correctly predicted the mortality status for 76% of the cases in the test set. The precision score, which measures the proportion of positive identifications that were actually correct, was 55%. This suggests that the model may have a relatively high rate of false positives. The recall score, which measures the proportion of actual positives that were correctly identified, was 77%. This is a significant metric given the high cost of false negatives in this context. The ROC-AUC score, which considers the trade-off between the true positive rate and false positive rate, was 77%. 

The model's performance suggests that it could be a useful tool for predicting mortality based on the ten selected features. However, the relatively low precision score indicates that there may be room for improvement, possibly by refining the feature selection process or changing the modelling approach to the machine learning techniques such as random forest classifier.

It is important to mention that the model's predictions are probabilistic, reflecting the uncertainty in predicting complex outcomes like mortality. This uncertainty is captured in the posterior distributions of the model parameters, which were estimated using MCMC sampling. While the model's performance is promising, further work could be done to improve its precision.
