```python
import matplotlib.pyplot as plt

def plot_histogram(dataframe, column_name):
    """
    Plot a histogram for a specific column in a pandas DataFrame.

    Parameters:
    - dataframe: pandas DataFrame
    - column_name: Name of the column to plot as a histogram
    """
    if column_name not in dataframe.columns:
        print(f"Column '{column_name}' not found in the DataFrame.")
        return

    plt.figure(figsize=(8, 6))
    plt.hist(dataframe[column_name], bins=40, color='blue', edgecolor='black')
    plt.title(f'Histogram of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon, norm, poisson, gamma
from scipy.optimize import curve_fit

def fit_and_plot_all_distributions(data, column_name, distributions, bins=20):
    column_data = data[column_name]

    def fit_and_plot_distribution(data, distribution, label=None, color=None):
        hist, bin_edges = np.histogram(data, bins=bins)
        bin_midpoints = (bin_edges[1:] + bin_edges[:-1]) / 2

        if distribution == "exponential":
            def pdf(x, lambd):
                return lambd * np.exp(-lambd * x)
            initial_guess = [1.0]
        elif distribution == "normal":
            def pdf(x, mu, sigma):
                return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
            initial_guess = [data.mean(), data.std()]
        elif distribution == "poisson":
            max_value = int(data.max())
            def pdf(x, mu):
                return poisson.pmf(x, mu)
            initial_guess = [max_value // 2]
        elif distribution == "gamma":
            def pdf(x, a, loc, scale):
                return gamma.pdf(x, a, loc, scale)
            initial_guess = [1.0, 0.0, 1.0]

        params, _ = curve_fit(pdf, bin_midpoints, hist, p0=initial_guess)
        params = params.ravel()

        plt.hist(data, bins=bins, edgecolor="black", density=True, alpha=0.7, label="Histogram")

        x_range = np.linspace(data.min(), data.max(), 1000)
        pdf_values = pdf(x_range, *params)
        plt.plot(x_range, pdf_values, color, label=label)

        plt.xlabel(column_name)
        plt.ylabel("Probability Density")
        plt.title(f"Histogram and Fitted {distribution.capitalize()} Distribution")

        plt.legend()
        plt.show()

    for dist_name, dist_label, dist_color in distributions:
        fit_and_plot_distribution(column_data, dist_name, label=dist_label, color=dist_color)
```

```python
distributions = [
    ("exponential", "Exponential", "r"),
    ("normal", "Normal", "g"),
    ("poisson", "Poisson", "c"),
    ("gamma", "Gamma", "m")
]
fit_and_plot_all_distributions(dataset, column_name, distributions)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, norm, uniform, poisson, lognorm

def fit_distribution_to_data(data, column_name, distribution, params):
    x_range = np.linspace(0, data[column_name].max(), 1000)
    pdf = None  # Initialize pdf variable
    pmf = None  # Initialize pmf variable

    if distribution == "exponential":
        loc, scale = 0, 1 / params  # Exponential:  = 1 / (sample mean)
        fitted_data = expon.rvs(loc=loc, scale=scale, size=len(data))
        pdf = expon.pdf(x_range, loc=loc, scale=scale)
        title = "Fitted Exponential Distribution"
    elif distribution == "normal":
        mu, sigma = params  # Normal: mean and standard deviation
        loc, scale = mu, sigma
        fitted_data = norm.rvs(loc=loc, scale=scale, size=len(data))
        pdf = norm.pdf(x_range, loc=loc, scale=scale)
        title = "Fitted Normal Distribution"
    elif distribution == "uniform":
        loc, scale = params  # Uniform: loc (minimum) and scale (range)
        fitted_data = uniform.rvs(loc=loc, scale=scale, size=len(data))
        pdf = uniform.pdf(x_range, loc=loc, scale=scale)
        title = "Fitted Uniform Distribution"
    elif distribution == "poisson":
        mu = params  # Poisson: mean parameter
        loc = 0
        fitted_data = poisson.rvs(mu, loc=loc, size=len(data))
        pmf = poisson.pmf(np.arange(data[column_name].min(), data[column_name].max() + 1), mu, loc=loc)
        title = "Fitted Poisson Distribution"
    elif distribution == "lognormal":
        s, loc, scale = params  # Log-Normal: shape (s), loc, and scale
        fitted_data = lognorm.rvs(s, loc=loc, scale=scale, size=len(data))
        pdf = lognorm.pdf(x_range, s, loc=loc, scale=scale)
        title = "Fitted Log-Normal Distribution"
    else:
        raise ValueError("Invalid distribution specified")

    return fitted_data, pdf, pmf, title

def fit_and_plot_distributions(data, column_name, distributions):
    x_range = np.linspace(0, data[column_name].max(), 1000)
    fitted_distributions = []
    for distribution, params in distributions:
        fitted_data, pdf, pmf, title = fit_distribution_to_data(data, column_name, distribution, params)
        fitted_distributions.append((fitted_data, pdf, pmf, title))

    # Plot all fitted distributions
    plt.figure(figsize=(15, 10))
    for i, (fitted_data, pdf, pmf, title) in enumerate(fitted_distributions, start=1):
        plt.subplot(2, 3, i)
        plt.hist(fitted_data, bins=20, density=True, alpha=0.7, edgecolor="black", label="Fitted Data")
        if distribution == "poisson":
            plt.plot(np.arange(data[column_name].min(), data[column_name].max() + 1), pmf, 'r', label="Fitted PMF")
        else:
            plt.plot(x_range, pdf, 'r', label="Fitted PDF")
        plt.xlabel(column_name)
        plt.ylabel("Probability Density")
        plt.title(title)
        plt.legend()

    plt.tight_layout()
    plt.show()

# Example usage:
# Assuming 'dataset' is your DataFrame and you want to analyze the 'Close' column
distributions = [
    ("exponential", 1 / dataset[column_name].mean()),
    ("normal", (dataset[column_name].mean(), dataset[column_name].std())),
    ("uniform", (dataset[column_name].min(), dataset[column_name].max() - dataset[column_name].min())),
    ("lognormal", (1, 0, np.exp(dataset[column_name].mean())))
    # ("poisson", dataset[column_name].mean()),

]

fit_and_plot_distributions(dataset, column_name, distributions)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson, expon, norm  # Import other supported distributions as needed
from scipy.optimize import curve_fit

def fit_and_plot_distribution(data, distribution, column_name, num_bins=20):
    # Select the specified column from the DataFrame
    column_data = data[column_name]

    # Prepare a list of supported distributions and their parameter estimators
    supported_distributions = {
        "poisson": (poisson, "mu", column_data.mean()),
        "exponential": (expon, "scale", 1 / column_data.mean()),  # Using the inverse for exponential rate
        "normal": (norm, "loc", column_data.mean()),
    }

    if distribution not in supported_distributions:
        raise ValueError(f"Unsupported distribution. Please choose from: {', '.join(supported_distributions.keys())}")

    # Get the distribution and parameter name
    dist, param_name, estimated_param = supported_distributions[distribution]

    # Fit the specified distribution to the data using the estimated parameter
    params = {param_name: estimated_param}
    fitted_data = dist.rvs(**params, size=len(column_data))

    # Create a histogram for the observed data and fitted data
    plt.figure(figsize=(8, 6))
    plt.hist(column_data, bins=num_bins, edgecolor="black", alpha=0.7, label=f"Observed {column_name} Data")
    plt.hist(fitted_data, bins=num_bins, edgecolor="black", alpha=0.5, label=f"Fitted {distribution.capitalize()} Distribution")
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.title(f"Fitting a {distribution.capitalize()} Distribution to {column_name} Data")
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

def fit_and_plot_all_distributions(dataset, column_name):
    # Prepare a list of supported distributions
    supported_distributions = ["poisson", "exponential", "normal"]  # Add more distributions as needed

    # Fit and plot each distribution
    for distribution in supported_distributions:
        fit_and_plot_distribution(dataset, distribution, column_name)

# Example usage:
# Assuming 'dataset' is your DataFrame and 'Rings' is the column name
fit_and_plot_all_distributions(dataset, column_name)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import expon, norm, uniform, poisson, gamma

def mom_estimator(data, distributions):
    estimators = {}

    for dist_name, dist_label in distributions:
        if dist_name == "exponential":
            sample_mean = data.mean()
            lambda_mom = 1 / sample_mean
            estimators[dist_label] = f"MOM Estimator for {dist_label}:  = {lambda_mom:.6f}"
        elif dist_name == "normal":
            sample_mean = data.mean()
            sample_std = data.std()
            estimators[dist_label] = f"MOM Estimators for {dist_label}: Mean = {sample_mean:.6f}, Standard Deviation = {sample_std:.6f}"
        elif dist_name == "uniform":
            estimators[dist_label] = f"MOM Estimators for {dist_label}: Minimum = {data.min()}, Maximum = {data.max()}"
        elif dist_name == "poisson":
            sample_mean = data.mean()
            estimators[dist_label] = f"MOM Estimator for {dist_label}:  = {sample_mean:.6f}"
        elif dist_name == "gamma":
            estimators[dist_label] = f"MOM Estimators for {dist_label}: Shape, Loc, and Scale Parameters - Not Applicable"
        else:
            estimators[dist_label] = f"Distribution '{dist_name}' not recognized. MOM Estimators are not applicable."

    return estimators
```

```python
from scipy.stats import expon, norm, uniform, poisson, lognorm, gamma
import numpy as np

def analyze_distributions(data, column_name, specific_event, distributions):
    results = {}
    
    for distribution, params in distributions:
        if distribution == "exponential":
            lambda_mom = params
            prob_gt_event = 1 - expon.cdf(specific_event, loc=0, scale=1 / lambda_mom)
            num_samples = 1000
            loc, scale = 0, 1 / lambda_mom
            random_samples = expon.rvs(loc=loc, scale=scale, size=num_samples)
            results[distribution] = (prob_gt_event, random_samples)
        elif distribution == "normal":
            mu, sigma = params
            prob_gt_event = 1 - norm.cdf(specific_event, loc=mu, scale=sigma)
            num_samples = 1000
            random_samples = norm.rvs(loc=mu, scale=sigma, size=num_samples)
            results[distribution] = (prob_gt_event, random_samples)
        elif distribution == "uniform":
            loc, scale = params
            prob_gt_event = 1 - uniform.cdf(specific_event, loc=loc, scale=scale)
            num_samples = 1000
            random_samples = uniform.rvs(loc=loc, scale=scale, size=num_samples)
            results[distribution] = (prob_gt_event, random_samples)
        elif distribution == "poisson":
            mu = params
            prob_gt_event = 1 - poisson.cdf(specific_event, mu)
            num_samples = 1000
            random_samples = poisson.rvs(mu, size=num_samples)
            results[distribution] = (prob_gt_event, random_samples)
        elif distribution == "lognormal":
            s, loc, scale = params
            prob_gt_event = 1 - lognorm.cdf(specific_event, s=s, loc=loc, scale=scale)
            num_samples = 1000
            random_samples = lognorm.rvs(s=s, loc=loc, scale=scale, size=num_samples)
            results[distribution] = (prob_gt_event, random_samples)
        elif distribution == "gamma":
            a, loc, scale = params
            prob_gt_event = 1 - gamma.cdf(specific_event, a=a, loc=loc, scale=scale)
            num_samples = 1000
            random_samples = gamma.rvs(a=a, loc=loc, scale=scale, size=num_samples)
            results[distribution] = (prob_gt_event, random_samples)
    
    return results
```

```python
# Example usage:
data = dataset[column_name]
specific_event = 0.1

distributions = [
    ("exponential", 1 / data.mean()),  # Exponential distribution with lambda_MOM
    ("normal", (data.mean(), data.std())),  # Normal distribution with mean and standard deviation
    ("uniform", (data.min(), data.max())),  # Uniform distribution with min and max
    ("poisson", data.mean()),  # Poisson distribution with lambda
    ("lognormal", (1, 0, np.exp(data.mean()))),  # Log-Normal distribution with shape, loc, and scale
    ("gamma", (2, 0, data.mean()))  # Gamma distribution with shape, loc, and scale
]

results = analyze_distributions(data, column_name, specific_event, distributions)

for distribution, (prob_gt_5, random_samples) in results.items():
    print(f"Distribution: {distribution}")
    print(f"Probability of X > {specific_event}: {prob_gt_5:.6f}")
    print(f"First 10 random samples: {random_samples[:10]}\n")
```
